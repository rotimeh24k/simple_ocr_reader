```python
# ocr = PaddleOCR(
#     det_model_dir=None,  # 문자 탐지 모델 (사용 안 할 경우 None)
#     rec_model_dir="./finetuned_model/",  # 학습된 인식 모델 경로
#     rec_char_type='ch',  # 언어 타입 (영어는 'en', 한글은 'ch')
#     use_angle_cls=False
# )

finetuned_model_path = "/content/output/ch_PP-OCR_V3_det/latest"

ocr = PaddleOCR(det_model_dir = './finetuned_model/',
                use_angle_cls=False,
                lang='korean',
                rec_model_dir=None,
                # 필요에 따라 설정 유지 또는 제거
                )
image_path = '../../data/book_cover_sample/images/책표지_총류_000002.jpg' # 이미지 파일 경로

# OCR 실행
result = ocr.ocr(image_path, cls=False)
print(result)
```

```python
# OCR 결과 출력
for line in result:
    for word_info in line:
        text, confidence = word_info[1]
        print(f"텍스트: {text}, 신뢰도: {confidence:.2f}")
```
